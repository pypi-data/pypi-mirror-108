# Recipe to automatize FT-ICR spectra phasing
a notebook to tune the algo, and see it running live...

in $r$ rather than in $f$

## Notations
- $\phi$ for signal phase
- $\psi$ for phase correction
    - $\phi - \psi = 0$
- signaux $n: 1 \cdots N$ , frequency: $f_n$ 
- $r_n = \frac{f_n-f_{min}}{\Delta f}$
- $\Delta f = f_{max}-f_{min}$

## problem
we assume

$ \phi_n = p_o + p_1 r_n + p_2 r^2_n$

where $p_o, p_1, p_2$ are the phase parameters. 
Corrections are noted $ P_o, P_1, P_2 = -p_o, -p_1, -p_2$ 

as D.K. mentioned, 2nd order polynomials are additive, and coefficients are additives.

we estimate $P_i \approx P^a_i + P^b_i + P^c_i$ in three steps

- from acquisition parameters: $P^a_1, P^a_2 $
- from small area autophasing $P^b_2$
- from a fit of peak phases $P^c_0 P^c_1 P^c_2$

# 2nd step
Estimate $P^b_2$ from $ \frac{d\phi}{dp}$ from spectrum

### estimate P0 , P1 on each peaks
For all peaks $n: 1 \cdots N$ estimate $\phi_n$ and $ \frac{d\phi}{dp}$ from spectrum

$$ \psi = P_o + P_1 r_n + P_2 r^2_n \quad \Rightarrow \frac{d\psi}{dp} = P_1 + 2 P_2 r$$

$$ \frac{d\psi}{dp} = 0 \quad \Rightarrow P_1 + 2 P_2 r = 0 \quad \Rightarrow r_c = \frac{-P_1}{2 P_2}$$
where $r_c$ is the position where local $p_1$ is zero *- crossing the axis -*

```python
def findpeaks(data, Dm, lowmassOI, highmassOI, noisethreshold, minimaldistance, baselinestrength, PH2, debug=False):
    """find peaks and holes
    data: the phase sensitive spectrum
    lowmassOI, highmassOI: defines region Of Interest,
    noisethreshold, minimaldistance: used by peak picker
    baselinestrength, PH2: used for baseline 
    """
    from spike.plugins.bcorr import autopoints
    Dm.set_unit('m/z').pp(autothresh=noisethreshold,  zoom=(highmassOI, lowmassOI))
    Dm.centroid()
    pks = peak_aggreg(Dm.peaks,minimaldistance, maxdist=3*minimaldistance)
    if debug: print(len(pks),'stored peaks')

    # find holes
    xp = autopoints(Dm, int(baselinestrength*PH2))
    if debug: print(len(xp),'baseline points')

    return (pks, xp)

def pkphase(D, pk, width=2, disp=False, dofirst=True):
    """phase 1st order around a peak, +/- width around in m/z"""
    # define zone around in points
    center = pk.pos
    largmz = D.axis1.itomz(center)
    oneDalton = (D.axis1.mztoi(largmz-1)-center) 
    zone = [int(center-width*oneDalton), int(center+width*oneDalton)]
    dd = spike.NPKData._NPKData(buffer=D.get_buffer()[zone[0]:zone[1]] )
    p0, p1 = 0, 0
    if disp:
        dd.display(label='before')
    dd.apmin(first_order=False)
    p0, p1 = p0+dd.axis1.P0, p1+dd.axis1.P1
    if dofirst:
        dd.apmin(first_order=True, bcpoints=8)
        p0, p1 = p0+dd.axis1.P0, p1+dd.axis1.P1
    if disp:
        plt.scatter(2*(center-zone[0]), pk.intens)
        dd.display(label='after',new_fig=False)
    return (    p0, p1, zone )

def fitP0P1(data, peaks, width=2):
    pl0 = []   # p0 over spectrum
    pl1 = []   # p1 over spectrum
    xpk = []   # coordinate of peak in pixel
    pksc = spike.plugins.Peaks.Peak1DList()  # peaks list cleaned
    for i in range(len(peaks)):
        try:
            p0,p1, zone = pkphase(data, peaks[i], width=width, disp=False)
        except:
            continue
        if p1 !=0:
            pc1 = p1 * data.cpxsize1/(zone[1]-zone[0])
            pl0.append(p0)
            pl1.append(pc1)
            xpk.append(peaks[i].pos)
            pksc.append(peaks[i])
    # return, xpk in 0..1, pl0 in degree, pl1 in turns
    return np.array(xpk)/data.cpxsize1, np.array(pl0), np.array(pl1)/360
```

### fit plc1 to  a line using robust stat

```python
def fitL1(xpeaks, p1list):
    "linear fit of p1list"
    from scipy.optimize import minimize
    def costL1(sol):
        "cost function implementing L1 norm"
        a,b = sol
        y = a*xpeaks+b
        return sum(abs(y-p1list))
    ini = (0,0)
    resL1 = minimize(costL1, ini)
    return resL1.x
pp2,pp1 = fitL1(xpk, pl1)
plt.plot(xpk,pp2*xpk+pp1, label='1st fit')
print(1, len(pl1),pp1,pp2)
```

### remove outliers and redo

```python
# Parameters
sigma_outliers = 1.0

n = 2
while True:
    res = (pp2*xpkc+pp1-plc1)
    res.mean(), res.std()
    xpkc2 = xpkc[abs(res)<sigma_outliers*res.std()]
    plc12 = plc1[abs(res)<sigma_outliers*res.std()]
    pp22,pp12 = fitL1(xpkc2, plc12)
    plt.plot(xpkc,pp22*xpkc+pp12, label='%dth fit'%n)
    plt.scatter(xpkc2, plc12, marker='+')
    plt.legend()
    print(n, len(plc12), pp12,pp22)
    if abs(pp2-pp22)<0.01*abs(pp2):
        break
    n += 1
    if n>6:
        break
    pp1,pp2 = pp12, pp22
```

### as a check, apply this correction

```python
Pb0 = 0
Pb1 = pp12
Pb2 = pp22/2
print('2nd correction', Pb0, Pb1, Pb2)
print('Correction Total', Pa0+Pb0, Pa1+Pb1, Pa2+Pb2)
print("r crossing:",-0.5*(Pa1+Pb1)/(Pa2+Pb2))
if abs(Pb2/Pa2)>0.5:
    print ("WARNING - correction is larger than 50% of initial guess")
    print("Try using Total as replacement of initial guess in step 1")
dd = D1.copy().phaseMS(Pb0, Pb1, Pb2, 0).display()
plt.gca().set_xbound(D1.axis1.mztoi(lowmassOI),D1.axis1.mztoi(highmassOI))  # to zoom in ROI
```

# 3rd step - Fine tune using 0th order.

### propagate correction to initial p0 values

```python
len(xpkc),len(pksc)
```

```python
newpl0 = (np.array(pl0)/360 - xpkc*(Pb1 + Pb2*xpkc))%1  # changed unit to turns for homogeneity
#newpl0 = residualphase
plt.figure()
plt.scatter(xpkc,newpl0,label='observed phase')
plt.ylabel('P0')
plt.legend()
plt.title('observed phase after 2nd step');
#################
pkph = []
pkphmoy = []
for i,pk in enumerate(pksc):
    ph = (-pkdispa(dd,pk,3)/360)%1
    pkph.append(ph)
    if abs(ph-newpl0[i])>0.7:
        if ph>newpl0[i]:
            moy = (0.5*(ph-1+newpl0[i]))%1
        else:
            moy = (0.5*(ph+1+newpl0[i]))%1
    else:
        moy = 0.5*(ph+newpl0[i])
    pkphmoy.append( moy )
pkphmoy = np.array(pkphmoy)
plt.scatter(xpkc,pkph, marker='+')
plt.scatter(xpkc,pkphmoy, marker='x')
```

### then unfold 

```python
# unfold phase v2
def unfold(phaselst):
    skip = phaselst[1:]-phaselst[:-1]
    for i in range(2,len(phaselst)):
        if skip[i-1]>0.75:
            phaselst[i:] -= 1
        if skip[i-1]<-0.75:
            phaselst[i:] += 1
unfold(newpl0)
plt.scatter(xpkc, newpl0, marker='+', label='unfolded')
plt.legend();
```

# and fit a 2nd order polynomial to it

```python
# first
final = np.polyfit(xpkc, newpl0, 2)
plt.plot(xpkc, np.polyval(final,xpkc),label='1st fit')

# reject outliers
n = 2
while True:
    res = newpl0-np.polyval(final,xpkc)
    xpkc2 = xpkc[abs(res)<2*res.std()]
    new2 = newpl0[abs(res)<2*res.std()]
    # redo
    final2 = np.polyfit(xpkc2, new2, 2)
    plt.plot(xpkc2, np.polyval(final2,xpkc2),label='%dth fit'%n)
    if abs(final[0]-final2[0])<0.001*abs(final[0]):
        break
    n += 1
    if n>6:
        break
    final = final2
    plt.scatter(xpkc2, new2, marker='x')

final = final2
plt.legend()
plt.title('2nd order fit of the residual phase, after unfolding')
print('P0 (degrees), P1, P2:', (final[2]*360)%360, final[1], final[0]*D1.cpxsize1)
```

```python
residualphase = (np.array(newpl0) - np.polyval(final,xpkc))
plt.figure()
plt.scatter(xpkc, 360*residualphase)
plt.ylabel('phase in degree')
plt.title('residual observed phase after 3rd step');
```

```python
Pc0, Pc1, Pc2 = (360*final[2]%360, final[1], final[0])
print('3rd correction', Pc0, Pc1, Pc2)
print('Correction Total', Pa0+Pb0+Pc0, Pa1+Pb1+Pc1, Pa2+Pb2+Pc2)

D1 = d1.copy() # copy the imported data-set to another object for processing
D1.apod_sin(maxi=0.05).zf(4).rfft()
D1.phaseMS(Pa0+Pb0+Pc0, Pa1+Pb1+Pc1, Pa2+Pb2+Pc2, 0).set_unit('m/z').real().display()
```

```python
def pkdispa(data, pk, extend):
    """
    data, a NPKDataset
    pk a Peak1dList
    extend: number of points (odd) on which the phase is evaluated
    """
    x = []
    y = []
    for i in range(-2*(extend//2),2*(extend//2)+1,2):
        x.append(data[2*int(round(pk.pos+i))])
        y.append(data[2*int(round(pk.pos+i))+1])
    return np.arctan2(sum(y),sum(x))*180/np.pi    
```
